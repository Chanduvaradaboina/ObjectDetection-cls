```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def apply_clahe(gray):
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    return clahe.apply(gray)
```

```python
def apply_bilateral(img):
    return cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
def dynamic_otsu(img):
    # Otsu automatically finds optimal threshold
    _, thresh = cv2.threshold(
        img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )
    return thresh
```

```python
def enhancement_pipeline(image):
    gray = image # The image is already grayscale, no need for cvtColor

    clahe_img = apply_clahe(gray)
    bilateral_img = apply_bilateral(clahe_img)
    otsu_img = dynamic_otsu(bilateral_img)

    return gray, clahe_img, bilateral_img, otsu_img
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Synthetic base image
base = np.zeros((256, 256), dtype=np.uint8)
cv2.circle(base, (128, 128), 60, 200, -1)
cv2.rectangle(base, (30, 30), (100, 100), 120, -1)
cv2.line(base, (0, 200), (255, 200), 180, 3)

def add_noise(image, noise_type):
    if noise_type == "gaussian":
        noise = np.random.normal(0, 25, image.shape)
        return np.clip(image + noise, 0, 255).astype(np.uint8)

    elif noise_type == "salt_pepper":
        noisy = image.copy()
        prob = 0.02
        rnd = np.random.rand(*image.shape)
        noisy[rnd < prob] = 0
        noisy[rnd > 1 - prob] = 255
        return noisy

    elif noise_type == "speckle":
        noise = np.random.randn(*image.shape)
        return np.clip(image + image * noise * 0.2, 0, 255).astype(np.uint8)

    elif noise_type == "poisson":
        return np.clip(np.random.poisson(image), 0, 255).astype(np.uint8)

    elif noise_type == "blur":
        return cv2.GaussianBlur(image, (9, 9), 2)

noise_types = ["gaussian", "salt_pepper", "speckle", "poisson", "blur"]

plt.figure(figsize=(12, 10))

for i, noise_type in enumerate(noise_types):
    noisy_img = add_noise(base, noise_type)

    plt.subplot(2, 3, i + 1)
    plt.imshow(noisy_img, cmap="gray")
    plt.title(f"Noisy ({noise_type})")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
titles = ["Grayscale", "CLAHE", "Bilateral Filtered", "Otsu Threshold"]

results = []
noise_types = ["gaussian", "salt_pepper", "speckle", "poisson", "blur"]

# Assuming 'base' image is available from previous cells (kernel state confirms this)
# Assuming 'add_noise' and 'enhancement_pipeline' functions are available (they are defined in previous cells)

for noise_type in noise_types:
    # Generate noisy image. 'base' variable is available from cell 'YBoEju9sLNU5'.
    noisy_img = add_noise(base, noise_type)
    # Apply the enhancement pipeline and collect the output
    processed_images = enhancement_pipeline(noisy_img)
    results.append(processed_images)

for i, res in enumerate(results):
    plt.figure(figsize=(12, 4))
    for j in range(4):
        plt.subplot(1, 4, j+1)
        plt.imshow(res[j], cmap='gray')
        plt.title(titles[j])
        plt.axis("off")
    plt.suptitle(f"Image {i+1} Enhancement Pipeline", fontsize=14)
    plt.show()
```
